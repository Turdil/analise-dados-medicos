```python
import pandas as pd
from google.colab import files
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np

```

```python
tb=pd.read_excel("dados_medicos_atualizado.xlsx")
desvio_pop=np.std(tb["Idade"],ddof=0)
desvio_amostral=np.std(tb["Idade"],ddof=1)
print(f'Desvio populacional: {desvio_pop:.2f}')
print(f'Desvio amostral: {desvio_amostral:.2f}')
```

```python
tb["Idade"]=tb["Idade"].fillna(tb["Idade"].median())
tb["Peso"]=tb["Peso"].fillna("Não informado")
display(tb)
```

```python
print(f'Média:{tb["Idade"].mean():.2f}')
print(f'Mediana:{tb["Peso"].median():.2f}')
```

```python
print(f'Diagnostico:')
print(tb["Diagnostico"].value_counts().plot(kind="bar", title="Diagnostico"))
plt.ylabel("Quantidade")
plt.xlabel("Diagnostico")
plt.grid(True)
plt.show()
```

```python
print (tb['Colesterol_mg_dl'].value_counts().plot(kind='bar',title="Colesterol_mg_dl"))
plt.xlabel("Colesterol")
plt.ylabel("Resultado")
plt.grid(True)
plt.show()
```
